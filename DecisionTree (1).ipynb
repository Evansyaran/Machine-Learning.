```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sal=pd.read_csv('/content/salaries (1).csv')
sal
```

```python
sal.columns
```

```python
inputs=sal.drop('salary_more_then_100k',axis='columns')
target=sal['salary_more_then_100k']
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

#Add new column to existing dataframe

```python
inputs['company_n']=le_company.fit_transform(inputs['company'])
inputs['job_n']=le_company.fit_transform(inputs['job'])
inputs['degree_n']=le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n=inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict_proba([[2,2,1]])
```

```python
model.predict([[2,2,1]])
```

###TiTanic Survival

```python
tit=pd.read_csv('/content/titanic (1).csv')
tit
```

```python
tit.columns
```

```python
tit=tit.drop(['PassengerId', 'Name', 'SibSp','Parch', 'Ticket',  'Cabin', 'Embarked'],axis='columns')
tit.head()
```

```python
dummies=pd.get_dummies(tit.Sex)
dummies.astype(int)
```

```python
tit_n=pd.concat([tit,dummies.astype(int)],axis='columns')
tit_n
```

```python
tit_n=tit_n.drop(['Sex'],axis='columns')
tit_n
```

```python
X=tit_n.drop(['Survived'],axis='columns')
X
```

```python
y=tit_n.Survived
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model1=DecisionTreeClassifier()
```

```python
model1.fit(X,y)
```

```python
model1.score(X,y)
```

```python
model1.predict([[3,20.0,100.2500,0,1]])
```
