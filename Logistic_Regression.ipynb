```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
insu=pd.read_csv('/content/insurance_data (1).csv')
insu
```

```python
insu.columns
```

```python
%matplotlib inline
plt.xlabel('age'),
plt.ylabel('bought_insurance')
plt.scatter(insu.age,insu['bought_insurance'],marker='+',color='blue')
```

```python
insu.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(insu[['age']],insu['bought_insurance'],test_size=0.1,random_state=10)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test)
```
